```python
# combining the three CDR datasets using concatenation

df_combine = pd.concat([df1,df2,df3], ignore_index = 1)
df_combine.head()
```

```python
# Dropping unnecessary columns 

df_combine_cleaned = df_combine.drop(columns = ['COUNTRY_A' ,	'COUNTRY_B' ,'CELL_ON_SITE'])
df_combine_cleaned.head()
```

```python
#combined analysis of the datasets both CDR and Cell geo
mtndata = pd.concat([df6, df_combine],ignore_index = 1)
mtndata.head()
```

```python
#drop missing values
mtndata_no_missing = mtndata.dropna()
mtndata_no_missing
```

```python
#checking for missing values
mtndata.isnull()
```

```python
# Drop rows where all cells in that row is NA
mtndata_cleaned = mtndata.dropna(how='all')
mtndata_cleaned
```

```python
# Drop column if they only contain missing values
mtndata.dropna(axis=1, how='all')
```

```python
# dropping duplicates in cells geo data
mtndata=mtndata.drop_duplicates(inplace=True)
```
